```python
import numpy as np
import pandas as pd

dataframe=pd.read_csv('MovieData.csv')
dataframe
```

```python
dataframe.info()
```

```python
#Drop rating, creative_type, source and production_method columns from database
newdataframe = dataframe.drop(columns=dataframe.columns[6:10])
newdataframe
```

```python
# check correct columns were dropped
newdataframe.info()
```

```python
# drop running_time columns from newdataframe
newestdataframe = newdataframe.drop(columns=newdataframe.columns[8:])
newestdataframe
```

```python
#check that only the last column was dropped
newestdataframe.info()
```

```python
print(newestdataframe)
```

```python
#Drop all columns after party affiliation
finaldataframe = newestdataframe.drop(columns=newestdataframe.columns[2])
finaldataframe
```
